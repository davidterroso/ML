# Distance Metrics

Before we start introducing the distances between some points, keep in mind that, to make the content more digestible, we are just going to demonstrate how to calculate the distance between points that are defined by two coordinates (e.g. the point A is defined by $(x_{A}, y_{A})$).

```python
import numpy as np
N_POINTS = 20
N_DIMS = 2
points = np.random.rand(N_POINTS, N_DIMS) * 10

indexes = np.random.choice(range(0, N_POINTS - 1), size=N_DIMS)
A, B = points[indexes[0], :], points[indexes[1], :]
```

## Distance in Real-Valued Vector Spaces

### Minkowski Distance <a id="minkowski-distance"></a>

The Minkowski Distance is the first distance presented because it encompasses two very important distances within itself, depending on the selected positive $p$ parameter. For two given points, $A (x_{A}, y_{A})$ and $B (x_{B}, y_{B})$, the Minkowski distance ($d_{M}$) is calculated as follows:

$$ d_{M} (A,B) = \sqrt[p]{\left(|x_{A} - x_{B}|\right)^{p} + \left(|y_{A} - y_{B}|\right)^{p}} $$

Just for further reference, this distance can be calculated for points that are defined by more than just two coordinates. With $F$ being the aggregate of all the coordinates/features that define points $A$ and $B$, this distance can be generalized to:

$$ d_{M} (A,B) = \sqrt[p]{\sum_{f \in F} \left(|f_{A} - f_{B}|\right)^{p}} $$

The fun part of the Minkowski Distance is the $p$ value. When we change this value to $p=1$, we get the Manhattan Distance, but when it is $p=2$, we get the Euclidean Distance.

```python
p = 1

# For the two dimensional case
# X_A, Y_A = A
# X_B, Y_B = B

# d_minkowski = abs(X_A - X_B) ** p + abs(Y_A - Y_B) ** p ) ** (1 / p)

# General, for the multi-dimensional case
d_minkowski = 0 
for i in range(N_DIMS):
    coord_diff = abs(A[i] - B[i]) ** p
    d_minkowski = d_minkowski + coord_diff 
d_minkowski = d_minkowski ** (1 / p)

print(d_minkowski)
```

### Manhattan, City Block, or L1 Distance <a id="manhattan-distance"></a>
The Manhattan Distance between two points $A$ and $B$, which can also be called City Block or L1 Distance, is therefore defined in the following equation:

$$ d_{L1} = |x_{A} - x_{B}| + |y_{A} - y_{B}| $$

```python
# For the two dimensional case
# X_A, Y_A = A
# X_B, Y_B = B

# d_manhattan = abs(X_A - X_B) + abs(Y_A - Y_B)

# General, for the multi-dimensional case
d_manhattan = 0 
for i in range(N_DIMS):
    coord_diff = abs(A[i] - B[i])
    d_manhattan = d_manhattan + coord_diff 

print(d_manhattan)
```

### Chebyshev Distance <a id="chebyshev-distance"></a>
Similarly to the Manhattan Distance, only the absolute difference between features is considered. However, while the Manhattan Distance is calculated as the sum of the differences between features, in the Chebyshev distance, only the largest difference between features is considered. Therefore, its expression can be written as:

$$ d_{C} = \text{max}(|x_{A} - x_{B}|, |y_{A} - y_{B}|) $$

```python
# For the two dimensional case
# X_A, Y_A = A
# X_B, Y_B = B

# d_chebyshev = max([abs(X_A - X_B), abs(Y_A - Y_B)])

# General, for the multi-dimensional case
list_coord_diff = []
for i in range(N_DIMS):
    coord_diff = abs(A[i] - B[i])
    list_coord_diff.append(coord_diff) 

d_chebyshev = max(list_coord_diff)

print(d_chebyshev)
```

### Euclidean or L2 Distance <a id="euclidean-distance"></a>
The Euclidean Distance is the most common distance used in KNN and other ML algorithms, being often referred to just as "distance". The Euclidean Distance between two points $A$ and $B$ is given by:

$$ d_{L2} = \sqrt{(x_{A} - x_{B})^{2} + (y_{A} - y_{B})^{2}} $$

A smoothing variable $V$ can also be used to calculate the Euclidean Distance, as follows:

$$ d_{SL2} = \sqrt{\frac{(x_{A} - x_{B})^{2} + (y_{A} - y_{B})^{2}}{V}} $$

```python
V = 1

# For the two dimensional case
# X_A, Y_A = A
# X_B, Y_B = B

# d_euclidean = ( ( (X_A - X_B) ** 2 + (Y_A - Y_B) ** 2 ) / V ) ** (1 / 2)

# General, for the multi-dimensional case
d_euclidean = 0
for i in range(N_DIMS):
    coord_diff = (A[i] - B[i]) ** 2
    d_euclidean = d_euclidean + coord_diff 

d_euclidean = (d_euclidean / V) ** (1 / 2)

print(d_euclidean)
```

#### NaN-Euclidean Distance <a id="nan-euclidean-distance"></a>
In ML, often times we deal with parameters that have missing values, where inputation may not always be an option, since the lack of knowledge in a feature can also, itself, be representative of a pattern. To workaround features with missing values, there is an adaptation of the Euclidean Distance in which the distance between two data points is only calculated between the coordinates with non-null values. This adaptation is called NaN-Euclidean-Distance and is defined as:

$$ d_{\text{NaN-L2}} (A,B) = \sqrt{\sum_{f \in F \bigcap f_{A} \notin \varnothing \bigcap f_{A} \notin \varnothing} \left(f_{A} - f_{B}\right)^{2}} $$

```python
V = 1

# For the two dimensional case
# X_A, Y_A = A
# X_B, Y_B = B

# d_euclidean = ( ( (X_A - X_B) ** 2 + (Y_A - Y_B) ** 2 ) / V ) ** (1 / 2)

# General, for the multi-dimensional case
d_euclidean = 0
for i in range(N_DIMS):
    if (A[i] is not None) and (B[i] is not None):
        coord_diff = (A[i] - B[i]) ** 2
        d_euclidean = d_euclidean + coord_diff 

d_euclidean = (d_euclidean / V) ** (1 / 2)

print(d_euclidean)
```

### Mahalanobis Distance <a id="mahalanobis-distance"></a>
The Mahalanobis Distance not only considers the difference between the coordinates between two data points, but also the covariance of each variable, and is calculated as:

$$ d_{\text{Mahalanobis}} = \sqrt{\left(A - B\right)^{T} V^{-1} \left(A - B\right)} $$

If we rewrite the Euclidean Distance equation to its vectorial form (as below), we can see the similarities between the Euclidean and the Mahalanobis distances, but also the differences between them.

$$ d_{L2} = \sqrt{\left(A - B\right)^{T} \left(A - B\right)} $$

The main difference between these distances is the Covariance Matrix, whose inverse is included in the Mahalanobis Distance expression. By including this matrix, the Mahalanobis distance is, unlike other distances, scale-invariant while still taking into account the correlations of the dataset. To better understand this matrix, let's see what it is all about. In this distance, let's consider the dataset of points $X$, composed of $m$ features and $n$ points, where each feature is denoted as $\left\{ x_{i} \right\}^{m}_{i=1}$. A general covariance matrix $V$ can be described as:

$$
V =
\begin{bmatrix}
    \text{Var}\left(x_{1}\right) & \dots  &  \text{Cov}\left(x_{m}, x_{1}\right) \\
    \vdots & \ddots & \vdots \\
    \text{Cov}\left(x_{1}, x_{m}\right) & \dots  &  \text{Var}\left(x_{m}\right) \\
\end{bmatrix}
$$

To make matters easier, let's start with a two-dimensional problem framing, where we have our usual two points A $(x_{1A}, x_{2A})$ and B $(x_{1B}, x_{2B})$, where each point has two features: $x_{1}$ and $x_{2}$. In the dataset $X$ with $m=2$ in which the points A and B are inserted, we assume that each coordinate ($x_{1}$ and $x_{2}$) is described by its own distribution that defines it, represented by $X_{1}$ and $X_{2}$, respectively. The distribution of $x_{1}$ and $x_{2}$ both have a standard deviation, represented by $\sigma_{1}$ and $\sigma_{2}$. When $n$ represents the total number of points in the dataset $X$, each of these values can be calculated as follows:

$$
\sigma_{1} = \sqrt{\frac{\left( \sum_{i \in X_{1}} x_{i1} - \bar{x}_{1} \right) ^ 2}{n-1}} = \sqrt{\frac{\left( \sum_{i \in X_{1}} \left( x_{i1} - \frac{1}{n} \sum_{i \in X_{1}} x_{i1} \right) \right) ^ 2}{n-1}}
$$
$$
\sigma_{2} = \sqrt{\frac{\left( \sum_{i \in X_{2}} x_{i2} - \bar{x}_{2} \right) ^ 2}{n-1}} = \sqrt{\frac{\left( \sum_{i \in X_{2}} \left( x_{i2} - \frac{1}{n} \sum_{i \in X_{2}} x_{i2} \right) \right) ^ 2}{n-1}}
$$

From here, we get: 

$$
\begin{aligned}
V &=
\begin{bmatrix}
    \text{Var}\left(x_{1}\right)  &  \text{Cov}\left(x_{2}, x_{1}\right) \\
    \text{Cov}\left(x_{1}, x_{2}\right)  &  \text{Var}\left(x_{2}\right) \\
\end{bmatrix} = \\
&=
\begin{bmatrix}
    \sigma_{1}^{2}  &  \sigma_{2}\sigma_{1} \\
    \sigma_{1}\sigma_{2}  &  \sigma_{2}^{2} \\
\end{bmatrix} = \\
&=
\begin{bmatrix}
    \sqrt{\frac{\left( \sum_{i \in X_{1}} \left( x_{i1} - \frac{1}{n} \sum_{i \in X_{1}} x_{i1} \right) \right) ^ 2}{n-1}}^{2}  &  \sqrt{\frac{\left( \sum_{i \in X_{2}} \left( x_{i2} - \frac{1}{n} \sum_{i \in X_{2}} x_{i2} \right) \right) ^ 2}{n-1}} \times \sqrt{\frac{\left( \sum_{i \in X_{1}} \left( x_{i1} - \frac{1}{n} \sum_{i \in X_{1}} x_{i1} \right) \right) ^ 2}{n-1}} \\
    \sqrt{\frac{\left( \sum_{i \in X_{1}} \left( x_{i1} - \frac{1}{n} \sum_{i \in X_{1}} x_{i1} \right) \right) ^ 2}{n-1}} \times \sqrt{\frac{\left( \sum_{i \in X_{2}} \left( x_{i2} - \frac{1}{n} \sum_{i \in X_{2}} x_{i2} \right) \right) ^ 2}{n-1}}  &  \sqrt{\frac{\left( \sum_{i \in X_{2}} \left( x_{i2} - \frac{1}{n} \sum_{i \in X_{2}} x_{i2} \right) \right) ^ 2}{n-1}}^{2} \\
\end{bmatrix} = \\
&=
\begin{bmatrix}
    \frac{\left( \sum_{i \in X_{1}} \left( x_{i1} - \frac{1}{n} \sum_{i \in X_{1}} x_{i1} \right) \right) ^ 2}{n-1}  &  \frac{ \left| \sum_{i \in X_{2}} \left( x_{i2} - \frac{1}{n} \sum_{i \in X_{2}} x_{i2} \right) \right| }{\sqrt{n-1}} \times \frac{ \left| \left( \sum_{i \in X_{1}} x_{i1} - \frac{1}{n} \sum_{i \in X_{1}} x_{i1} \right) \right| }{\sqrt{n-1}} \\
    \frac{ \left| \left( \sum_{i \in X_{1}} x_{i1} - \frac{1}{n} \sum_{i \in X_{1}} x_{i1} \right) \right| }{\sqrt{n-1}} \times \frac{ \left| \left( \sum_{i \in X_{2}} x_{i2} - \frac{1}{n} \sum_{i \in X_{2}} x_{i2} \right) \right| }{\sqrt{n-1}}  &  \frac{\left( \sum_{i \in X_{2}} \left( x_{i2} - \frac{1}{n} \sum_{i \in X_{2}} x_{i2} \right) \right) ^ 2}{n-1} \\
\end{bmatrix} = \\
&=
\begin{bmatrix}
    \frac{\left( \sum_{i \in X_{1}} \left( x_{i1} - \frac{1}{n} \sum_{i \in X_{1}} x_{i1} \right) \right) ^ 2}{n-1}  &  \frac{ \left| \left( \sum_{i \in X_{2}} x_{i2} - \frac{1}{n} \sum_{i \in X_{2}} x_{i2} \right) \right|  \times  \left| \sum_{i \in X_{1}} \left( x_{i1} - \frac{1}{n} \sum_{i \in X_{1}} x_{i1} \right) \right| }{n-1} \\
    \frac{ \left| \sum_{i \in X_{1}} \left( x_{i1} - \frac{1}{n} \sum_{i \in X_{1}} x_{i1} \right) \right| \times \left| \sum_{i \in X_{2}} \left( x_{i2} - \frac{1}{n} \sum_{i \in X_{2}} x_{i2} \right) \right| }{n-1}  &  \frac{\left( \sum_{i \in X_{2}} \left( x_{i2} - \frac{1}{n} \sum_{i \in X_{2}} x_{i2} \right) \right) ^ 2}{n-1} \\
\end{bmatrix}
\end{aligned}
$$

After computing the covariance matrix, we can finally calculate the Mahalanobis distance. Because we are dividing the differences between points by their covariance, the distance is not dependent on any scale, which is useful when the points are of different orders of magnitude. You can also think of the covariance matrix the matrix multiplication between the array that contains the standard deviation of all features $\vec{\sigma}$, defined as $\vec{\sigma} = [\{\sigma_{i}\}_{i=1}^{m}]$, with shape $(1, m)$. If you perform the matrix multiplication $\vec{\sigma}^{T} \cdot \vec{\sigma}$, you get a $m \times m$ matrix, which corresponds to your covariance matrix.

```python
import numpy as np

def get_std(arr: list) -> float:
    if arr is not None:
        arr_sum, std = [0] * 2
        for i in range(len(arr)):
            arr_sum = arr_sum + arr[i]
        arr_mean = arr_sum / len(arr)

        for i in range(len(arr)):
            std = std + (arr[i] - arr_mean) ** 2
        std = ( std / ( len(arr) - 1 ) ) ** ( 1 / 2 )
        # std = ( std / len(arr) ) ** ( 1 / 2 ) # NumPy calculates the Standard Deviation 
                                                # without the Bessel's Correction
                                                # the -1 term in the denominator that is 
                                                # set to correct the bias and considers 
                                                # the mean as the first degree of freedom
                                                # Hence, the results are different below 
        # print("Manually calculated std: ", std)
        # print("Automatically calculated std: ", np.std(arr))
        return float(std)

stds = []
for dim in range(N_DIMS):
    stds.append(get_std(list(points[:,dim])))

cov_mat = np.array(stds)[:, None] @ np.array(stds)[:, None].T

print(stds)
```

## Distance in Two-Dimensional Real-Valued Vector Spaces

### Haversine Distance <a id="haversine-distance"></a>
Unlike the previous distances, the Haversine Distance can only be calculated between two points in a sphere, given their longitude and latitude (in radians). Consequently, the distance between the two data points is also in radians. Obviously, the Haversine Distance is not your usual distance between data points in any feature space, being limited to data points in which the features correspond to coordinates. Nevertheless, it is calculated as follows, for two points $A$ and $B$ defined, respectively, as $(\phi_{A}, \lambda_{A})$ and $(\phi_{B}, \lambda_{B})$:

$$ d_{H} = 2R \times \text{arcsin}\left( \sqrt{\sin^{2}\left( \frac{\phi_{B} - \phi_{A}}{2} \right) + \cos (\phi_{1}) \cos(\phi_{2}) \sin^{2}\left( \frac{\lambda_{B} - \lambda_{A}}{2} \right)} \right)$$

Since the $R$ variable is the radius of the sphere, when we want to calculate the distance between two points on Earth the value of $R$ would be $6371 \text{km}$, for example.
